# Exploratory Data Analysis - Netflix Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/netflix_data.csv')

df.head()
```

```
df.info()
df.describe()
```

```
genre_counts = df['Genre'].value_counts()
print(genre_counts)

genre_counts.plot(kind='bar', title='Genre Distribution')
plt.show()
```

```
type_counts = df['Type'].value_counts()

type_counts.plot(kind='pie', autopct='%1.1f%%', title='Content Type Distribution')
plt.show()
```

```
year_trend = df['Release_Year'].value_counts().sort_index()

year_trend.plot(title='Content Release Trend')
plt.show()
```

```
country_counts = df['Country'].value_counts()

country_counts.head(10).plot(kind='bar', title='Top Countries')
plt.show()
```
